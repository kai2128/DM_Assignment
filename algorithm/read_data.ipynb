```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import metrics
```

```python
original_data = pd.read_csv('./vw.csv')
data_encoded = pd.read_csv('./vw_encoded.csv')
data_encoded.drop(['Unnamed: 0'], axis=1, inplace=True)
data_encoded
```

### Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_encoded.drop(columns=['price'])
y = np.ravel(data_encoded[['price']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def regression_report(y_test, y_pred):
    print('R-squared score: \t{:>14.4f}'.format(
        metrics.r2_score(y_test, y_pred)))
    print('Mean Absolute Error(MAE): \t{:.4f}'.format(
        metrics.mean_absolute_error(y_test, y_pred)))
    print('Mean Squared Error (MSE): \t{:.4f}'.format(
        metrics.mean_squared_error(y_test, y_pred)))
    print('Root Mean Squared Error (RSME): {:.4f}'.format(
        np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
```
